```python
# Task 2: Mark the success/failed launches for each site on the map
# Create a new map centered at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Define color for success/failure
def get_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'  # Successful launches
    else:
        return 'red'    # Failed launches

# Add circles and markers for each launch
for index, row in spacex_df.iterrows():
    launch_coordinate = [row['Lat'], row['Long']]
    # Determine the color based on the launch outcome
    color = get_color(row['class'])
    
    # Create a circle for each launch
    circle = folium.Circle(
        launch_coordinate,
        radius=100,
        color=color,
        fill=True
    ).add_child(folium.Popup(f"Launch Site: {row['Launch Site']}\nOutcome: {'Success' if row['class'] == 1 else 'Failure'}"))
    
    # Create a marker for each launch
    marker = folium.Marker(
        launch_coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:{color};"><b>{"Success" if row['class'] == 1 else "Failure"}</b></div>',
        )
    )
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
